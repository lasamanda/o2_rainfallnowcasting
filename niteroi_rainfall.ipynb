# Plotando gráfico de chuvas de 1976-1995

```python
df76_95 = df76_95.drop(labels=0)
df76_95
```

```python
#Preparando x e y
x = pd.to_datetime(df76_95['Data'], format='%d/%m/%Y')

df76_95['Total'] = df76_95['Total'].str.replace(',', '.').astype(float)
y = df76_95['Total']
```

```python
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, y)
plt.xlabel("Ano")
plt.ylabel("Precipitação total (mês)")
plt.title("Total de precipitação por mês de 1976-1995")
```

De acordo com o gráfico, o total precipitado mensamental não mostrou tendencia a aumentar durante o período observado.

## Somando total anual

```python
df_anual = df76_95[['Data','Total']]
df_anual = df_anual.drop(labels=1, axis=0)
df_anual["Data"] = pd.to_datetime(df_anual["Data"], format='%d/%m/%Y')
df_anual
```

```python
total_anual = df_anual.groupby(df_anual['Data'].dt.year)['Total'].sum()
total_anual
```

```python
fig, ax = plt.subplots(figsize=(12,6))

plt.plot(total_anual.index, total_anual.values)
plt.xlabel("Ano")
plt.ylabel("Precipitação total anual")
plt.title("Precipitação total anual 1976-1994")

# Note that 1995 is still included in the graph despite being deleted of the dataset
```
